```python
import numpy as np
import pandas as pd
import random
import names
import datetime as dt, dateutil
from dateutil.relativedelta import relativedelta
```

```python
text = pd.read_csv("1.csv")
text["Tag"].unique()
```

```python
df = pd.read_html("https://www.medicalhomeportal.org/diagnoses-and-conditions/diagnosis-prevalence-list")
diseases = []
for i in range(len(df[0])):
    if len(df[0]['Sort by Diagnosis'][i].split()) <= 2:
        diseases.append(df[0]['Sort by Diagnosis'][i])

insurances = ["Western","Eastern","Northern","Southern","Pacific"]
insurance = random.choice(insurances)
```

```python
def gen_date(start_date,end_date):
    time_between_dates = end_date - start_date
    days_between_dates = time_between_dates.days
    random_number_of_days = random.randrange(days_between_dates)
    random_date = (start_date + dt.timedelta(days=random_number_of_days))
    b_date_gen = random_date.strftime('%m/%d/%Y')
    return b_date_gen

gen_date(dt.date(1940, 1, 1), dt.date(2000, 2, 1))
```

```python
for files in range(50000):
    text = pd.read_csv("1.csv")

    #String

    patient_first_name = text[(text["Tag"] == "I-PATIENT NAME")].Text.to_string(index=False).strip()
    patient_last_name = text[(text["Tag"] == "B-PATIENT NAME")].Text.to_string(index=False).strip()
    phy_first_name = text[(text["Tag"] == "B-ADMITTING PHYSICIAN")].Text.to_string(index=False).strip()
    phy_last_name = text[(text["Tag"] == "I-ADMITTING PHYSICIAN")].Text.to_string(index=False).strip()
    b_diagnosis = text[(text["Tag"] == "B-ADMITTING DIAGNOSIS")].Text.to_string(index=False).strip().split('\n')[0]
    i_diagnosis = text[(text["Tag"] == "I-ADMITTING DIAGNOSIS")].Text.to_string(index=False).strip().split('\n')[0]
    b_insurance = text[(text["Tag"] == "B-PRIMARY INSURANCE PLAN")].Text.to_string(index=False).strip()

    #Numerical

    b_account = text[(text["Tag"] == "B-ACCOUNT #")].Text.to_string(index=False).strip()
    b_mr = text[(text["Tag"] == "B-MR NUMBER")].Text.to_string(index=False).strip()
    b_ss = text[(text["Tag"] == "B-SOCIAL SECURITY #")].Text.to_string(index=False).strip()
    i_ss = text[(text["Tag"] == "I-SOCIAL SECURITY #")].Text.to_string(index=False).strip()
    b_patient_phone = text[(text["Tag"] == "B-PATIENT PHONE #")].Text.to_string(index=False).strip()

    #Date

    admit_date = text[(text["Tag"] == "B-ADMIT DATE")].Text.to_string(index=False).strip()
    b_date = text[(text["Tag"] == "B-BIRTH DATE")].Text.to_string(index=False).strip()
    b_age = text[(text["Tag"] == "B-AGE")].Text.to_string(index=False).strip()

    text.replace(to_replace = patient_first_name, value = names.get_first_name(), inplace = True)   
    text.replace(to_replace = patient_last_name, value = names.get_last_name(), inplace = True)
    text.replace(to_replace = phy_first_name, value = names.get_first_name(), inplace = True)
    text.replace(to_replace = phy_last_name, value = names.get_last_name(), inplace = True)

    insurance = random.choice(insurances)
    text.replace(to_replace = b_insurance, value = random.choice(insurances), inplace = True)

    disease = random.choice(diseases)
    if len(disease.split()) == 2:
        text.replace(to_replace = b_diagnosis, value = disease.split()[0], inplace = True)
        text.replace(to_replace = i_diagnosis, value = disease.split()[1], inplace = True)
    else:
        text.replace(to_replace = b_diagnosis, value = disease.split()[0], inplace = True)
        text.drop(text.index[text['Tag'] == "I-ADMITTING DIAGNOSIS"], inplace = True)

    text.replace(to_replace = b_account, value = random.randint(10000,99999), inplace = True)
    text.replace(to_replace = b_mr, value = random.randint(10000000,99999999), inplace = True)

    ss_index = text[(text["Tag"] == "B-SOCIAL SECURITY #")].index.values[0]
    text.replace(to_replace = text["Text"][ss_index], value = random.randint(100,999), inplace = True)
    text.replace(to_replace = text["Text"][ss_index+2], value = random.randint(10,99), inplace = True)
    text.replace(to_replace = text["Text"][ss_index+4], value = random.randint(1000,9999), inplace = True)

    text.replace(to_replace = b_patient_phone, value = random.randint(1000000000,9999999999), inplace = True)

    text.replace(to_replace = b_date, value = gen_date(dt.date(1930, 1, 1), dt.date(2010, 1, 1)), inplace = True)

    text.replace(to_replace = admit_date, value = gen_date(dt.date(2010, 2, 1), dt.date(2020, 2, 1)), inplace = True)


    from datetime import datetime
    admit_date = text[(text["Tag"] == "B-ADMIT DATE")].Text.to_string(index=False).strip()
    b_date = text[(text["Tag"] == "B-BIRTH DATE")].Text.to_string(index=False).strip()

    admit_date = datetime.strptime(admit_date, '%m/%d/%Y')
    b_date = datetime.strptime(b_date, '%m/%d/%Y')
    text.replace(to_replace = b_age, value = relativedelta(admit_date,b_date).years, inplace = True)

    text.to_csv(str(files)+"_augmented.csv",index=False)
```
